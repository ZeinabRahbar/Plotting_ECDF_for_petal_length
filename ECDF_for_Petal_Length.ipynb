```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Read the dataset
data = pd.read_csv("iris.csv")
data.head()
```

```python
data.columns
```

# ECDF for petal_length

```python

# Extract the 'sepal_length' column
petal_length = data['petal_length']

# Calculate the ECDF
x = np.sort(petal_length)
y = np.arange(1, len(x) + 1) / len(x)

# Plot the ECDF
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Set the plot labels and title
plt.xlabel('petal Length')
plt.ylabel('ECDF')
plt.title('ECDF Plot for petal Length')

# Show the plot
plt.show()
```
